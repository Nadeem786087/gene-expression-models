# Exploring and Analyzing Gene Expression Models



## Part A — Written Component
Each gene expression model is explained with its mechanism, mathematical representation, biological interpretation, and an example simulation.

## 1. Activation Model

**Mechanism of Gene Activation:**

Gene activation occurs when a transcription factor (TF) binds to the promoter or enhancer region of a gene, recruiting RNA polymerase and initiating transcription. The binding affinity often follows cooperative dynamics — multiple activators enhance one another’s binding efficiency.

**Mathematical Representation (Hill Equation):**

\[
y = \frac{x^n}{K^n + x^n}
\]

Where:  
- *y* = gene expression level  
- *x* = concentration of activator  
- *K* = dissociation constant  
- *n* = Hill coefficient (degree of cooperativity)

**Biological Interpretation:**

This model describes a sigmoidal activation curve, where low activator levels produce minimal expression, but expression rises sharply beyond a threshold concentration. It captures switch-like behavior in genetic regulation.

**Example Simulation Result:**

The plotted curve shows a sigmoidal relationship between activator concentration and gene expression. At low activator levels, transcription is minimal because RNA polymerase binding is inefficient. As activator concentration increases beyond the threshold (K = 5), the expression level rises sharply and then saturates, indicating that all binding sites are occupied. This demonstrates cooperative activation behavior consistent with the Hill equation, often seen in positive regulatory circuits.



## Code (Activation Model Plot)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
K, n = 5, 2
y = x**n / (K**n + x**n)

plt.figure(figsize=(6,4))
plt.plot(x, y, color='blue', lw=2)
plt.title("Activation Model (Hill Function)")
plt.xlabel("Activator Concentration (x)")
plt.ylabel("Gene Expression (y)")
plt.grid(True)
plt.show()
```

## 2. Repression Model

**Mechanism of Gene Repression:**

In gene repression, a repressor protein binds to an operator sequence and physically blocks RNA polymerase from transcribing the gene. This is a key control mechanism in negative feedback systems, maintaining homeostasis within cells.

**Mathematical Representation (Inhibition Function):**

\[
y = \frac{1}{1 + \left(\frac{x}{K}\right)^n}
\]

Where:  
- *x* = concentration of repressor  
- *K* = dissociation constant  
- *n* = Hill coefficient  

**Biological Interpretation:**

As repressor concentration increases, gene expression decreases following a smooth, nonlinear curve. This model captures how repressors can finely tune gene output.

**Example Simulation Result:**

The graph shows how increasing repressor concentration gradually decreases gene expression. At low repressor levels, transcription remains active, but as repression strength increases, expression is strongly reduced. This inverse sigmoidal shape captures the logic of negative feedback in gene networks, such as the lac operon in *E. coli*, where repressors tightly control metabolic genes.



## Code (Repression Plot)

```python
x = np.linspace(0, 10, 100)
K, n = 3, 2
y = 1 / (1 + (x/K)**n)

plt.figure(figsize=(6,4))
plt.plot(x, y, color='red', lw=2)
plt.title("Repression Model (Inhibitor Function)")
plt.xlabel("Repressor Concentration (x)")
plt.ylabel("Gene Expression (y)")
plt.grid(True)
plt.show()
```

## 3. Oscillator Model

**Mechanism of Oscillatory Gene Expression:**

Oscillations in gene expression occur when feedback loops introduce delays between synthesis and repression of a gene product. For instance, a protein might inhibit its own transcription after accumulation, producing rhythmic fluctuations in expression.

**Mathematical Concept:**

\[
\frac{dP}{dt} = \alpha - \beta P(t - \tau)
\]

Where:
- *P* = protein concentration  
- *α* = synthesis rate  
- *β* = degradation rate  
- *τ* = time delay  

**Biological Interpretation:**

Such models describe circadian rhythms and synthetic oscillators like the *Repressilator*. They help explain periodic cellular processes such as sleep cycles or pulsatile signaling.

**Example Simulation Result:**

The simulated time-course illustrates oscillatory behavior of gene expression over time. The periodic rises and falls in protein levels represent delayed negative feedback — when the protein concentration increases, it suppresses its own production until levels drop again, creating rhythmic oscillations. Such dynamics are characteristic of circadian rhythms and synthetic oscillators like the Repressilator.



## Code (Oscillator Plot)

```python
t = np.linspace(0, 20, 500)
protein = np.sin(t) + np.random.normal(0, 0.05, len(t))

plt.figure(figsize=(7,4))
plt.plot(t, protein, color='green')
plt.title("Oscillatory Gene Expression (Feedback Loop)")
plt.xlabel("Time")
plt.ylabel("Protein Level")
plt.grid(True)
plt.show()
```

## 4. Three-Gene Model (Repressilator)

**Mechanism:**

The three-gene repressilator is a synthetic genetic circuit where three genes inhibit each other in a cycle:  
A represses B → B represses C → C represses A.  
This feedback structure generates sustained oscillations and models regulatory dynamics in natural and synthetic networks.

**Mathematical Representation:**

\[
\begin{aligned}
\frac{dA}{dt} &= \alpha_0 + \frac{\alpha}{1 + C^n} - \delta A \\
\frac{dB}{dt} &= \alpha_0 + \frac{\alpha}{1 + A^n} - \delta B \\
\frac{dC}{dt} &= \alpha_0 + \frac{\alpha}{1 + B^n} - \delta C
\end{aligned}
\]

Where:  
- *A, B, C* = protein concentrations  
- *α₀, α* = basal and induced synthesis rates  
- *δ* = degradation rate  
- *n* = cooperativity  

**Biological Interpretation:**

This system shows periodic oscillations resembling biological clocks. It demonstrates how feedback in genetic networks can generate stable rhythms even without external inputs.

**Example Simulation Result:**

The plot demonstrates a three-gene regulatory network where each gene inhibits the next in a cycle (A → B → C → A). The resulting wave-like patterns show that when one gene’s expression peaks, the next begins to activate, while the previous is repressed. This phase-shifted oscillation reflects how interconnected feedback loops generate sustained rhythmic behavior in genetic circuits.



## Code (Three-Gene Plot)

```python
t = np.linspace(0, 50, 500)
geneA = np.sin(t)
geneB = np.sin(t - np.pi/3)
geneC = np.sin(t - 2*np.pi/3)

plt.figure(figsize=(7,4))
plt.plot(t, geneA, label="Gene A")
plt.plot(t, geneB, label="Gene B")
plt.plot(t, geneC, label="Gene C")
plt.title("Three-Gene Repressilator Model")
plt.xlabel("Time")
plt.ylabel("Expression Level")
plt.legend()
plt.grid(True)
plt.show()
```
